## Q1 (i)
For an arbitrary point in $\mathbb{R}^2$ represented in polar coordinates $(r, \theta)$, define $f$ as follows:

For $r \geq 2$, we have:

$$
f(r, \theta) = 
\begin{cases}
			r & 0 \leq \theta \leq 2\pi (\frac{r-2}{r}),\\
            r(r -1 - \frac{r \theta}{2 \pi}) & 2\pi (\frac{r-2}{r}) < \theta < 2\pi (\frac{r-1}{r}),\\
			r^2 (\frac{\theta}{2 \pi} - \frac{r-1}{r}) & 2\pi (\frac{r-1}{r}) \leq \theta \leq 2\pi
		 \end{cases}
$$

For $r < 2$, 

$$
f(r, \theta) = r f(2, \theta)
$$

A diagram of $f(r, \theta)$ is given below for an arbitrary $r \geq 2$. Note that this function is constantly equal to $r$ until $\theta = \frac{r-2}{r} (2 \pi)$, then dips in a straight line to $0$ at $\theta = \frac{r-1}{r} (2 \pi)$, then comes back to $r$ at $\theta = 2 \pi$ in a straight line. The idea is that we squeeze the dip to $0$ ever closer to $2 \pi$ without ever reaching it. Note that $f$ is continuous in $\theta$ for a given $r$, and for any $\theta \in [0, 2 \pi)$, $\lim_{r \to r_1} f(r, \theta) = f(r_1, \theta)$ so the function is continuous. The definition for $r < 2$ ensures that $f$ is globally continuous, as we smoothly deform it to the origin.

To show that $f$ is not coercive, define a sequence of points in $\mathbb{R}^2$ in polar coordinates as follows:

$$
\textbf{x}_n := (r_n, \theta_n) = (n, \frac{n-1}{n} (2 \pi)) \ \ \  \ \ \forall n \in \mathbb{N}
$$

Then by the definition of $f$ above, we have that $f(\textbf{x}_n) = 0 \ \forall n \in \mathbb{N}$ but $\lVert \textbf{x}_n \rVert \to \infty$ as $n \to \infty$, therefore $f$ is not coercive.

<br>
<br>

Now to prove that $f$ satisfies the required property, first note that for any fixed $\theta \in [0, 2 \pi)$, $\lim_{r \to \infty} f(r, \theta) = \infty$, since for $r$ large enough we have that $\frac{r-2}{r} (2 \pi) > \theta \implies f(r, \theta) = r$ by the definition above. This means that $f$ **eventually goes to infinity in all directions**, which is the required property. 

To show this, take an arbitrary $\alpha \in \mathbb{R}$. If $\alpha \neq 0$, for a point $(x_1, \alpha x_1)$ in Cartesian coordinates, this is just the point $(|x_1|\sqrt{1 + \alpha^2}, \arctan(\alpha))$ in polar coordinates. Now:

$$
\lim_{|x_1| \to \infty} f(|x_1|\sqrt{1 + \alpha^2}, \arctan(\alpha)) = \infty
$$

Since the angle $\arctan(\alpha) \in [0, 2 \pi)$ is fixed.

Similarly, for $(\alpha x_2, x_2)$ in Cartesian coordinates, this is $(|x_2|\sqrt{1 + \alpha^2}, \arctan(\frac{1}{\alpha}))$ in polar coordinates. Since the angle $\arctan(\frac{1}{\alpha}) \in [0, 2 \pi)$ is fixed, we similarly have that

$$
\lim_{|x_2| \to \infty} f(|x_2|\sqrt{1 + \alpha^2}, \arctan(\frac{1}{\alpha})) = \infty
$$

So for $\alpha \neq 0$, we have that 

$$
\lim_{|x_1| \to \infty} f(x_1, \alpha x_1) = \lim_{|x_2| \to \infty} f(\alpha x_2, x_2) = \infty
$$

(where the above is in cartesian coordinates).

Now if $\alpha = 0$, $(x_1, 0)$ is simply the point $(|x_1|, 0)$ or $(|x_1|, \pi)$ in polar coordinates. Similarly, the point $(0, x_2)$ is simply the point $(|x_2|, \frac{\pi}{2})$ or $(|x_2|, \frac{3 \pi}{2})$ in polar coordinates. Therefore again as each point has that $\theta$ is fixed and $r \to \infty$,

$$
\lim_{|x_1| \to \infty} f(|x_1|, 0) = \lim_{|x_1| \to \infty} f(|x_1|, \pi) = \infty
$$
and
$$
\lim_{|x_2| \to \infty} f(|x_2|, \frac{\pi}{2}) = \lim_{|x_2| \to \infty} f(|x_2|, \frac{3 \pi}{2}) = \infty
$$

So for any $\alpha \in \mathbb{R}$,

$$
\lim_{|x_1| \to \infty} f(x_1, \alpha x_1) = \lim_{|x_2| \to \infty} f(\alpha x_2, x_2) = \infty
$$

(in cartesian coordinates) which is the required property.

\nabla^2 f(\lambda, 2\lambda^2) =

           32\lambda^2 & -8\lambda \\

$$
[ x \ \ \ y ] \begin{bmatrix}

         \end{bmatrix}

## Q2 (i)

We can re-write this dynamical process as:

$$
x_0 = \bar{x}
$$
$$
x_1 = a \bar{x} + du_1
$$
$$
x_2 = a^2 \bar{x} + adu_1 + du_2
$$
$$
...
$$
$$
x_i = a^i\bar{x} + a^{i-1}du_1 + ... + adu_{i-1} + du_i
$$
Which can be rewritten in matrix form:

$$
\textbf{x} = S \textbf{u} - \textbf{b} 
$$
where $S \in \mathbb{R}^{(N+1) \times N}$

Where 
$$
S = 
(d)\begin{bmatrix}
           0 & 0 & 0 & \cdots & 0 & 0\\
           1 & 0 & 0 & \cdots & 0 & 0\\
           a & 1 & 0 & \cdots & 0 & 0\\
           a^2 & a & 1 & \cdots & 0 & 0\\
           \vdots & \vdots & \vdots & \ddots & \vdots & \vdots\\
           a^{N-1} & a^{N-2} & a^{N-3} & \cdots & 1 & 0 \\
           a^{N} & a^{N-1} & a^{N-2} & \cdots & a & 1
         \end{bmatrix}
$$
and
$$
\textbf{b} = -\bar{x}
\begin{bmatrix}
         1 \\
         a \\
         a^2 \\
        \vdots \\
        a^N

         \end{bmatrix}
$$
Then (DO) takes the form:
$$
\min\limits_{\textbf{u} \in \R^N} \ \lVert S \textbf{u} - \textbf{b} \rVert^2_2 + \frac{\gamma}{2} \lVert \textbf{u} \rVert^2_2
$$

Which is precisely the regularised least squares problem for $\textbf{u}$ seen in the notes, with $\lambda = \frac{\gamma}{2}$ and $D = I_N$.

Thus, from the notes, the optimisation problem admits a unique solution iff $ \text{Null}(S) \cap \text{Null}(D) = \{ \textbf{0} \}$. Therefore, since $\text{Null}(I_N) = \{ \textbf{0} \}$, we necesserily have this condition, which means the system has a unique solution given by:

$$
\textbf{u}_{RLS} = (S^TS + \gamma D^T D)^{-1}S^T \textbf{b} = (S^T S + \frac{\gamma}{2} I_N)^{-1} S^T \textbf{b}
$$

Now take any $\textbf{u}$ solving the unregularised least squares problem, and $\textbf{u}^*$ solving the regularised least squares problem, we have by the definition of minimality that:

$$
\lVert S\textbf{u} - \textbf{b} \rVert \leq \lVert S\textbf{u}^* - \textbf{b} \rVert
$$
and
$$
\lVert S\textbf{u}^* - \textbf{b} \rVert + \frac{\gamma}{2}\lVert \textbf{u}^* \rVert \leq \lVert S\textbf{u} - \textbf{b} \rVert + \frac{\gamma}{2}\lVert \textbf{u} \rVert
$$
$$
\implies \frac{\gamma}{2}\lVert \textbf{u}^* \rVert \leq \frac{\gamma}{2}\lVert \textbf{u} \rVert
$$
$$
\implies \lVert \textbf{u}^* \rVert \leq \lVert \textbf{u} \rVert
$$

## Q2 (ii)

```python
N = 40
a = 1
d = -0.01
x_bar = 1
gammas = [1e-3, 1e-2, 0.1, 1]

# Get optimal control signal u_star

# Get S
S = np.zeros((N+1,N))
for i in range(N):
    for j in range(i+1):
        S[i+1][j] = d * a ** j
#print(S)

# Get b
b = np.zeros(N+1)
for i in range(len(b)):
    b[i] = -x_bar * a**i

# For each gamma, plot optimal control and trajectory
for gamma in gammas:

    # Get RLS solution
    u_rls = np.matmul(np.matmul(np.linalg.inv(np.matmul(np.transpose(S), S) + gamma * np.identity(N)), np.transpose(S)), b)

    # Get optimal trajectory
    x_u = np.matmul(S, u_rls) - b

    # Plot
    plt.title(f"Plot for gamma = {gamma}")
    plt.plot(u_rls, label='Control signal')
    plt.plot(x_u, label='Trajectory')
    plt.legend()
    plt.show()
```

Increasing $\gamma$ makes the optimal trajectory decay to $0$ much slower, as the control signal is flattened due to the high penalty for a high value of $\lVert \textbf{u} \rVert^2_2$. This is because the $\frac{\gamma}{2} \lVert \textbf{u} \rVert^2_2$ term aims to reduce the size of $\textbf{u}$ and make it more flat, since higher values of $u_i$ are penalised much more due to this being the squared norm. (For example the squared norm of $(15, 0, 0)$ is larger than $(5, 5, 5)$ so this term tends to make $\textbf{u}$ more flat). On the other hand, the $\lVert \textbf{x} \rVert^2_2$ term is minimised by a control signal that is very sharp at the beginning to bring the trajectory down as quickly as possible and keep it low. So the solution is a tradeoff between these two behaviours, with a higher gamma term prioritising the penalty in the control signal, thus making the control signal more flat and with a lesser peak at the start.

## Q2 (iii)

Here to find the optimal control signal with the extra penalty function I use gradient descent. Define our function $f(\textbf{x})$ to minimise as follows:

$$
f(\textbf{u}) = \ \lVert S \textbf{u} - \textbf{b} \rVert^2_2 + \frac{\gamma}{2} \lVert \textbf{u} \rVert^2_2 - \delta \sum_{i=1}^{N} \log (u_{max} - u_i)
$$

Now note that 
$$
\lVert S \textbf{u} - \textbf{b} \rVert^2_2 = (S \textbf{u} - \textbf{b})^T(S \textbf{u} - \textbf{b}) = \textbf{u}^T S^T S \textbf{u} - \textbf{b}^T S \textbf{u} - (S\textbf{u})^T \textbf{b} + \textbf{b}^T \textbf{b}
$$
And since $\textbf{b}^T S \textbf{u}$ is a scalar and therefore equals its transpose, we have $\textbf{b}^T S \textbf{u} = (S\textbf{u})^T \textbf{b}$ so we have:
$$
\lVert S \textbf{u} - \textbf{b} \rVert^2_2 = \textbf{u}^T S^T S \textbf{u} - 2(\textbf{b}^T S) \textbf{u} + \textbf{b}^T \textbf{b}
$$
which is simply a quadratic function for $\textbf{u}$, for which we know the derivative.

Similarly, we have $\frac{\gamma}{2} \lVert \textbf{u} \rVert^2_2 = \frac{\gamma}{2} \textbf{u}^T \textbf{u}$, another quadratic function.

Now to take the derivative of our function $f$, we first consider the derivative of $\delta \sum_{i=1}^{N} \log (u_{max} - u_i)$:

$$
\frac{\partial}{\partial u_i} (\delta \sum_{i=1}^{N} \log (u_{max} - u_i)) = \frac{-\delta}{u_{max} - u_i}
$$
So if we define the vector $\boldsymbol \beta (\textbf{u}) \in \mathbb{R}^n$ to have entries $\frac{\delta}{u_{max} - u_i}$ for each entry $i$, then we get the gradient of our function $f$:

$$
\nabla f(\textbf{u}) = 2 S^T S \textbf{u} - 2 S^T \textbf{b} + \gamma \textbf{u} + \boldsymbol \beta (\textbf{u})
$$
by using the formula for gradient of a quadratic function in the notes.

The implementation for this is given below:

My initial guess is $(\frac{u_{max}}{2}, \frac{u_{max}}{2}, ..., \frac{u_{max}}{2})^T$, and I use a constant stepsize of $\frac{1}{10}$, which has yielded very good convergence.

```python
N = 40
a = 1
d = -0.01
x_bar = 1
u_max = 8
gamma = 1e-2
delta = 1e-2
# Get S
S = np.zeros((N+1,N))
for i in range(N):
    for j in range(i+1):
        S[i+1][j] = d * a ** j
# Get b
b = np.zeros(N+1)
for i in range(len(b)):
    b[i] = -x_bar * a**i


# Define f, our loss function
def f(u, S, b, u_max, gamma, delta):
    """
    Loss function for our optimisation problem. Returns a scalar given a u vector and the parameters.
    """
    return np.linalg.norm(np.matmul(S, u) - b)**2 + gamma/2 * (np.linalg.norm(u)**2) - delta * sum(np.log(u_max - u))

# Define derivative of f
def grad_f(u, S, b, u_max, gamma, delta):
    """
    Gradient function for the loss function f. Returns nabla f(u) for any vector u.
    """
    return 2 * np.matmul(np.matmul(np.transpose(S), S), u) - 2 * np.matmul(np.transpose(S), b) + gamma*u + delta / (u_max - u)

# Implement gradient descent for optimal u
u_0 = u_max / 2 * np.ones(N) # Initial guess

# Do gradient descent, plotting loss and optimal trajectory and control signal.
loss = []
for i in range(10000): # Do 10000 iterations

    gradient = grad_f(u_0, S, b, u_max, gamma, delta)
    u_0 = u_0 - gradient/10 # Stepsize of gradient /10
    loss.append(f(u_0, S, b, u_max, gamma, delta))

plt.title("Loss function after n iterations")
plt.plot(loss)
plt.show()

plt.title("Control signal and trajectory for this problem")
plt.plot(u_0, label='Control signal')
plt.plot(np.matmul(S, u_0) - b, label='Trajectory')
plt.axhline(u_max, color='black', label='u_max')
plt.legend()
plt.show()

# Show that gradient is very close to 0, indicating we have found a local minimum.
print("---")
print("Maximum entry in the gradient vector after all epochs:")
print(np.max(grad_f(u_0, S, b, u_max, gamma, delta)))
print("---")
```

Comments: I have used gradient descent with a constant stepsize. Observations:
- The gradient at the final iteration is very close to 0, indicating we have a found a (local) minimum for the loss funciton $f$, so the optimisation method has worked.
- The control signal, similarly to Q2 (ii), looks very high at the start and then decays fast, which is the behaviour generated by the $\lVert S \textbf{u} - \textbf{b} \rVert^2_2 $ and $ \frac{\gamma}{2} \lVert \textbf{u} \rVert^2_2$ terms. However, in this optimisation problem the signal is forced to stay below $u_{max}$ for each $u_i$, since the gradient becomes infinitely steep as we approach $u_{max}$ from any direction. We can see this pressure to stay below the $y=8$ line in the plot, as each $u_i$ is below $u_{max}$.


## Q2 (iv)

The $\mathcal{L}_{\epsilon}$ is a regulariser as it penalises vectors $\textbf{u}$ with low sparsity when added to the loss function. The function is indeed differentiable, as it is piecewise differentiable and the different components of the function coincide both in value and derivative at $u_i = \pm \epsilon$. The derivative is given by:

$$
l_{\epsilon}(u_i) := \frac{d}{du_i} \mathcal{L}_{\epsilon} (u_i) = 

\begin{cases}
			2u_i & \text{if $|u_i| \leq \epsilon$}\\
            \text{sgn}(u_i)\epsilon & \text{otherwise.}
		 \end{cases}
$$

where $\text{sgn}$ is the sign function.

Now our function $f$ to minimise is:
$$
f(\textbf{u}) = \ \lVert S \textbf{u} - \textbf{b} \rVert^2_2 + \frac{\gamma_2}{2} \lVert \textbf{u} \rVert^2_2 + \gamma_1 \sum_{i=1}^{N} \mathcal{L}_{\epsilon} (u_i)
$$
And if we define the vector $\textbf{l}_\epsilon (\textbf{u}) \in \mathbb{R}^N$ to be the vector with $i$ th entry $l_{\epsilon}(u_i)$ we have that the gradient is given by:
$$
\nabla f(\textbf{u}) = 2 S^T S \textbf{u} - 2 S^T \textbf{b} + \gamma_2 \textbf{u} + \gamma_1 \textbf{l}_\epsilon (\textbf{u})
$$
for any $\textbf{u} \in \mathbb{R}^N$.

We now implement a gradient descent method with backtracking to solve this optimisation problem:

Settings: $s = \frac{1}{10}, \alpha = 0.8, \beta = 0.5$.

```python
s = 1/10
alpha = 0.8
beta = 0.5

N = 40
a = 1
d = -0.01
x_bar = 1
epsilon = 3

# Get S
S = np.zeros((N+1,N))
for i in range(N):
    for j in range(i+1):
        S[i+1][j] = d * a ** j
# Get b
b = np.zeros(N+1)
for i in range(len(b)):
    b[i] = -x_bar * a**i


def L_epsilon(u):
    """
    Returns the value of L_epsilon on the whole vector u
    """
    result = []
    for i in range(len(u)):
        if np.abs(u[i]) <= epsilon:
            result.append((u[i]**2) / 2)
        else:
            result.append(epsilon * (np.abs(u[i]) - epsilon/2))
    
    return np.array(result)


# Define f, our loss function
def f(u, S, b, gamma1, gamma2):
    """
    Loss function for our optimisation problem. Returns a scalar given a u vector and the parameters.
    """
    return np.linalg.norm(np.matmul(S, u) - b)**2 + gamma2/2 * (np.linalg.norm(u)**2) + gamma1 * sum(L_epsilon(u))

# Define derivative of f
def grad_f(u, S, b, gamma1, gamma2):
    """
    Gradient function for the loss function f. Returns nabla f(u) for any vector u.
    """
    # Construct l_epsilon for this u
    l_epsilon_u = []
    for i in range(len(u)):
        if np.abs(u[i]) <= epsilon:
            l_epsilon_u.append(2 * u[i])
        else:
            l_epsilon_u.append(epsilon * np.sign(u[i]))

    l_epsilon_u = np.array(l_epsilon_u)

    return 2 * np.matmul(np.matmul(np.transpose(S), S), u) - 2 * np.matmul(np.transpose(S), b) + gamma2*u + gamma1 * l_epsilon_u

u_0 = 2 * np.ones(N) # Initial guess


# IMPLEMENT GRADIENT DESCENT WITH BACKTRACKING

# ------------------------------------------------------------
# CASE i) GAMMA1 = 0, GAMMA2 = 10^-2
gamma1 = 0
gamma2 = 1e-2
loss = []
for i in range(10000): # Do 10000 iterations

    gradient = grad_f(u_0, S, b, gamma1, gamma2)

    # Backtracking step
    t_k = s

    while f(u_0, S, b, gamma1, gamma2) - f(u_0 - t_k*gradient, S, b, gamma1, gamma2) < -alpha * t_k * np.matmul(np.transpose(gradient), gradient):
        t_k = beta * t_k

    u_0 = u_0 - gradient * t_k # Dynamic stepsize
    loss.append(f(u_0, S, b, gamma1, gamma2))

plt.title("Case i)")
plt.plot(u_0, label='Control signal')
plt.plot(np.matmul(S, u_0) - b, label='Trajectory')
plt.legend()
plt.show()



# ------------------------------------------------------------
# CASE ii) GAMMA1 = 10^-2, GAMMA2 = 0
gamma1 = 1e-2
gamma2 = 0
loss = []
for i in range(10000): # Do 10000 iterations

    gradient = grad_f(u_0, S, b, gamma1, gamma2)

    # Backtracking step
    t_k = s

    while f(u_0, S, b, gamma1, gamma2) - f(u_0 - t_k*gradient, S, b, gamma1, gamma2) < -alpha * t_k * np.matmul(np.transpose(gradient), gradient):
        t_k = beta * t_k

    u_0 = u_0 - gradient * t_k # Dynamic stepsize
    loss.append(f(u_0, S, b, gamma1, gamma2))

plt.title("Case ii)")
plt.plot(u_0, label='Control signal')
plt.plot(np.matmul(S, u_0) - b, label='Trajectory')
plt.legend()
plt.show()
```

Above we can see the plot for case i): $\gamma_1 = 0$ and $\gamma_2 = 10^{-2}$ and case ii): $\gamma_1 = 10^{-2}$ and $\gamma_2 = 0$.

Case i) is simply the optimisation problem in Q2 (ii) as $\gamma_1 = 0$ so the graph looks the same as above (backtracking has converged to the RLS optimal solution). Case ii) penalises the control signal $\textbf{u}$ only with the $\mathcal{L}_{\epsilon}$ which is akin to the $1$-norm. Thus, control signals which are spread out are not more favoured than ones where there is a sharper spike but the same mass, like was the case with case i). This leads to a very sharp spike at the beginning for case ii) as the trajectory is brought down as quickly as possible to minimise the value of $\lVert \textbf{x}_{\bar{x}}^{\textbf{u}} \rVert ^2_2$ whilst keeping $\textbf{u}$ sparse. Thus the optimal trajectory decays quicker for case ii).
